```python
# importing needed modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import local cvs file from my computer
movies = pd.read_csv("C:\\Users\\abdul\\Dropbox\\PC\\Desktop\\IS362\\movies.csv", index_col = 0)  
movies
```

```python
# 1. Replacing NaN value with 0
movies2 = movies.fillna(0.0)
movies2
```

```python
# 2a. Finding the Average Movie Ratings by user
Ratings = movies2
Ratings.mean(axis = 1).round(2) # axis 1 is Y axis
```

```python
# 2b. Finding the Average Ratings of Movie by movies
Ratings.mean(axis = 0).round(2)
```

```python
# 3a. Normalized Data
normalized = (Ratings - Ratings.min())/ (Ratings.max()- Ratings.min())
normalized
```

```python
# 3b. Normalized data in average
print(normalized.mean(axis=0))
print(normalized.mean(axis=1))
```

```python
# 4: Normalized ratings are a way of transforming ratings data into a more consistent and 
# comparable form. This can be useful for a variety of tasks, such as recommending products to
# users or comparing the performance of different products.
```

```python
# 5a. EXTRA CREDIT:
standardized = (Ratings - Ratings.mean())/ Ratings.std()
standardized
```

```python
# 5b. Standarized data in numbers
print(standardized.mean(axis=0))
print(standardized.mean(axis=1))
```
